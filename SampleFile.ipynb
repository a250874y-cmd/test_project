```python
import numpy as np
import matplotlib.pyplot as plt

# set x[0]=0, x[1]=1/999*2*pi,...,x[999]=999/999*2*pi
x = np.linspace(0, 2 * np.pi, 1000)

# calculate y1=sin(x)
y1 = np.sin(x)

# calculate y2=cos(2 * x)
y2 = np.cos(2 * x)

# convert x to degree
x_deg = x * 180.0 / np.pi

# draw the figure
plt.plot(x_deg, y1, 'r', x_deg, y2, 'b')
```
